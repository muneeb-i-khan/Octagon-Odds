```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('ufc.csv')
df
```

```python
df = df.drop([df.index[0]])
df
```

```python
df['Win%'] = df['W'] / (df['W'] + df['L'] + df['D']) *100
df
```

```python
df = df[df['Reach'] != '--']
df
```

```python
df.reset_index(drop=True, inplace=True)
df
```

```python
df = df.drop(columns=['First', 'Last', 'Nickname'])
df
```

```python
df.dropna(subset=['Stance'], inplace=True)
df
```

```python
df.rename(columns={'Ht.': 'Height', 'Wt.' : 'Weight'}, inplace=True)
df
```

```python
df.drop(columns=['Belt'], inplace = True)
df
```

```python
grouped_df_ht = df.groupby('Height')['Win%'].mean()
grouped_df_ht
```

```python
plt.figure(figsize=(10, 6))  
plt.plot(grouped_df_ht.index, grouped_df_ht.values, marker='o', linestyle='-', color='b')
plt.xlabel('Height')  
plt.ylabel('Win%')  
plt.title('Win % vs. Height')  
plt.grid(True)  
plt.show()  
```

```python
grouped_df_wt = df.groupby('Weight')['Win%'].mean()
grouped_df_wt
```

```python
plt.figure(figsize=(10, 6))
plt.plot(grouped_df_wt.index, grouped_df_wt.values, marker='o', linestyle='-', color='b')
plt.xlabel('Weight')
plt.ylabel('Win%')
plt.title('Win % vs. Weight')
plt.grid(True)

plt.xticks(rotation=45)

plt.show()
```

```python
grouped_df_reach = df.groupby('Reach')['Win%'].mean()
grouped_df_reach
```

```python
plt.figure(figsize=(10, 6))
plt.plot(grouped_df_reach.index, grouped_df_reach.values, marker='o', linestyle='-', color='b')
plt.xlabel('Reach')
plt.ylabel('Win%')
plt.title('Win % vs. Reach')
plt.grid(True)

plt.xticks(rotation=45)

plt.show()
```

```python
grouped_df_stance = df.groupby('Stance')['Win%'].mean()
grouped_df_stance
```

```python
plt.figure(figsize=(10, 6))
plt.bar(grouped_df_stance.index, grouped_df_stance.values, color='b', alpha=0.7)  # Use plt.bar() instead of plt.plot()
plt.xlabel('Stance')
plt.ylabel('Win%')
plt.title('Win % vs. Stance')
plt.grid(True)

plt.xticks(rotation=45) 
plt.show()
```

```python
grouped_df_wt = grouped_df_wt.reset_index()
grouped_df_wt
```

```python
def extract_integer(weight_str):
    return int(''.join(filter(str.isdigit, weight_str)))

grouped_df_wt['Weight'] = grouped_df_wt['Weight'].apply(extract_integer)
grouped_df_wt
```

```python
grouped_df_ht = grouped_df_ht.reset_index()
grouped_df_ht
```

```python
import re

def height_to_cm(height_str):
    pattern = r'(\d+)\'\s*(\d+)"' 
    match = re.match(pattern, height_str)
    if match:
        feet, inches = map(int, match.groups())
        total_inches = feet * 12 + inches
        return total_inches * 2.54
    else:
        return None  

grouped_df_ht['Height'] = grouped_df_ht['Height'].apply(height_to_cm)

grouped_df_ht
```

```python
grouped_df_reach = grouped_df_reach.reset_index()
grouped_df_reach
```

```python
def remove_suffix(s):
    return s[:-3] if s.endswith(".0\"") else s


grouped_df_reach['Reach'] = grouped_df_reach['Reach'].apply(remove_suffix)

grouped_df_reach
```
